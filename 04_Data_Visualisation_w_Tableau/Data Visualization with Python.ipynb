```python
plt.plot(x,y, 'g-')
#plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)   # *args :  istedigin sayida parametre kullanacagini belirtmek icin. 
plt.plot(x,y,'r')   # More on color options later  # *args geriye tuple(demet) veri tipini döndürürken
                     # **kwargs ise paremetremiz sozluk yapisinda ise istedigimiz kadar kullanabilecegimizi belirtmek icin.
plt.subplot(1,2,2)   # **kwargs yapisinda (sayi1=5,sayi2=6,sayi3=7) 
plt.plot(y,x,'g*-')    # *kwargs dictionary(sözlük) veri tipini döndürdü.
```

```python

# Add set of axes to figure
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
axes1.plot(x,y)
axes1.set_title('BIGGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')


axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2), dpi=100)
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```
