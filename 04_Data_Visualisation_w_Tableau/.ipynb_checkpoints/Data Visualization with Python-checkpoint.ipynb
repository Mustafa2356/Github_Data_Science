```python
plt.plot(x,y, 'g-')  # x ekseninde x verileri y ekseninde y verileri g- green yesil ve cizgili grafik
#plt.show()
plt.xlabel('X Label') 
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)   # *args :  istedigin sayida parametre kullanacagini belirtmek icin. 
plt.plot(x,y,'r')   # More on color options later  # *args geriye tuple(demet) veri tipini döndürürken
                     # **kwargs ise paremetremiz sozluk yapisinda ise istedigimiz kadar kullanabilecegimizi belirtmek icin.
plt.subplot(1,2,2)   # **kwargs yapisinda (sayi1=5,sayi2=6,sayi3=7) 
plt.plot(y,x,'g*-')    # *kwargs dictionary(sözlük) veri tipini döndürdü.
                    # subplot icerisindeki ilk rakam yani 1 satiri ikinci rakam 2 stunu 
                    # 3. rakam ise iki tane alt grafikten 1. sini ve digeride 2. sini belirtiyor
```

```python
# Add set of axes to figure
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
axes1.plot(x,y)
axes1.set_title('BIGGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')


axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
## Functional Method
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Title')
```

```python
## OO Method
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
fig, axes = plt.subplots()
axes.plot(x, y)
axes.set_xlabel('X Axis Title')
axes.set_ylabel('Y Axis Title')
axes.set_title('Title')
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 3, figsize=(12, 4))
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[1].plot(x, y)
axes[1].set_title("title")  # dizinler 0 ,1, 2 seklinde gidiyor biz 1. birinci dizindeki grafigi cizdirdik
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
fig, axes = plt.subplots(2, 2, figsize=(8, 4))
axes[0,1].plot(x, y)
axes[0,1].set_title("title") # sifirinci dizindeki satir uzerinden birinci dizin cizdirdik
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, y, color="red", lw=3, ls='-', marker='o', markersize=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # left, bottom, width, height (range 0 to 1)
ax.plot(x,y,color='purple', lw=1, ls='-')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
